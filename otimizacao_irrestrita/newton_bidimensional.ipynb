```python
e = 0.00001

#Função de teste: f(x, y) = 5x² - 5xy + 2.5y² - x - 1.5y
def f(x, y):

  return 5 * x * x - 5 * x * y + 2.5 * y * y - x - 1.5 * y

#Derivada numérica em relação a x da função de teste f
def dfdx(x, y):

  return (f(x + e, y) - f(x - e, y)) / (2 * e) 

#Derivada numérica em relação a y da função de teste f
def dfdy(x, y):

  return (f(x, y + e) - f(x, y - e)) / (2 * e) 

#Derivada segunda numérica em relação a x da função de teste f
def dfdxx(x, y):

  return (dfdx(x + e, y) - dfdx(x - e, y)) / (2 * e) 
  
#Derivada segunda numérica em relação a y da função de teste f
def dfdyy(x, y):

  return (dfdy(x, y + e) - dfdy(x, y - e)) / (2 * e) 

#Derivada segunda numérica em relação a x e y da função de teste f
def dfdxy(x, y):

  return (dfdx(x, y + e) - dfdx(x, y - e)) / (2 * e) 
```

```python
#Vetor gradiente da função de teste f calculado numericamente
def gradiente(x, y):

  return np.matrix([[dfdx(x, y)], [dfdy(x, y)]])
```

```python
#Matriz hessiana da função de teste f calculada numericamente
def hessiana(x, y):

  return np.matrix([[dfdxx(x, y), dfdxy(x, y)], [dfdxy(x, y), dfdyy(x, y)]])
```

```python
#Cálculo da matriz inversa
def inversa(m):

  den = m[0, 0] * m[1, 1] - m[0, 1] * m[1, 0]

  if den == 0:

    return np.matrix([[0, 0], [0, 0]])

  return np.matrix([[m[1, 1], -m[0, 1]], [-m[1, 0], m[0, 0]]]) / den
```

```python
#Otimização utilizando o método de Newton para funções bidimensionais
def newton(x, y, t):

  while True:

    i = inversa(hessiana(x, y))
    g = gradiente(x, y)

    if round(g[0,0], 10) == 0 and round(g[1,0], 10) == 0:

      break

    d = -i * g

    x = x + t * d[0,0]
    y = y + t * d[1,0]

  return x, y, f(x,y)
```

```python
#Testando o algoritmo do método de Newton bidimensional para a função f
import numpy as np

x = 10
y = 20
t = 0.75

x_otimo, y_otimo, f_otimo = newton(x, y, t)

print("O valor ótimo foi encontrado em x =", x_otimo, "e y =", y_otimo, "onde f(x, y) =", f_otimo)
```
