```python
#Função de teste: f(x) = 2sen(x) - x^2/10
def f(x):
  
  return 2 * math.sin(x) - x * x / 10
```

```python
#Função que procura o ponto ótimo (máximo) da função f pela busca da razão áurea. Para buscar o mínimo, basta inverter a primeira desigualdade abaixo
def aurea(lim_inf, lim_sup, max_i, erro):
  
  phi = (5 ** 0.5 - 1) / 2
  d = phi * (lim_sup - lim_inf)
  x1 = lim_inf + d
  x2 = lim_sup - d
  f1 = f(x1)
  f2 = f(x2)
  i = 0
  
  while True:
    
    if f1 > f2:
      
      lim_inf = x2
      d = phi * (lim_sup - lim_inf)
      x2 = x1
      f2 = f1
      x1 = lim_inf + d
      f1 = f(x1)
      x_otimo = x1
      f_otimo = f1
        
    else:
      
      lim_sup = x1
      d = phi * (lim_sup - lim_inf)
      x1 = x2
      f1 = f2
      x2 = lim_sup - d
      f2 = f(x2)
      x_otimo = x2
      f_otimo = f2
      
    i += 1

    if x_otimo != 0:
    
      if i == max_i or (1 - phi) * abs((lim_sup - lim_inf) / x_otimo) * 100 <= erro :
        
        break

  return x_otimo, f_otimo
```

```python
#Testando o algoritmo da busca da razão áurea para a função f
import math

lim_inf = 0
lim_sup = 4
max_i = 30
erro = 0.001

x_otimo, f_otimo = aurea(lim_inf, lim_sup, max_i, erro)

print("O valor ótimo foi encontrado em x =", x_otimo, "onde f(x) =", f_otimo)
```
