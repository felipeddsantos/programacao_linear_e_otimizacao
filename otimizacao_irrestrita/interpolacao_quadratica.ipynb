```python
#Função de teste: f(x) = 2sen(x) - x^2/10
def f(x):

  return 2 * math.sin(x) - x * x / 10
```

```python
#Função que procura o ponto ótimo (máximo) da função f por interpolação quadrática. Para buscar o mínimo, basta inverter a primeira desigualdade abaixo
def interpolacao(x0, x1, x2, max_i):

  i = 0

  while True:

    num = f(x0) * (x1 * x1 - x2 * x2) + f(x1) * (x2 * x2 - x0 * x0) + f(x2) * (x0 * x0 - x1 * x1)
    den = 2 * f(x0) * (x1 - x2) + 2 * f(x1) * (x2 - x0) + 2 * f(x2) * (x0 - x1)

    if den == 0 or i == max_i:

      break

    x3 = num / den

    if f(x3) < f(x1):
      
      x_otimo = x1
      f_otimo = f(x1)
      x2 = x1
      x1 = x3
    
    else:
      
      x_otimo = x3
      f_otimo = f(x3)
      x0 = x1
      x1 = x3
    
    i += 1

  return x_otimo, f_otimo
```

```python
#Testando o algoritmo da interpolação quadrática para a função f
import math

x0 = 0
x1 = 1
x2 = 2
max_i = 30

x_otimo, f_otimo = interpolacao(x0, x1, x2, max_i)

print("O valor ótimo foi encontrado em x =", x_otimo, "onde f(x) =", f_otimo)
```
