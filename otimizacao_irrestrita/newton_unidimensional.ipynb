```python
#Função de teste: f(x) = 2sen(x) - x^2/10
def f(x):

  return 2 * math.sin(x) - x * x / 10

#Derivada da função de teste: f'(x) = 2cos(x) - x/5
def df(x):

  return 2 * math.cos(x) - x / 5

#Derivada segunda da função de teste: f''(x) = -2sen(x) - 1/5
def df2(x):

  return -2 * math.sin(x) - 1 / 5
```

```python
#Função que procura o ponto ótimo da função f pelo método de Newton
def newton(x0, max_i):

  i = 0

  while True:

    num = df(x0)
    den = df2(x0)

    if den == 0 or i == max_i:

      break

    x = x0 - num / den

    x0 = x
    i += 1

  return x0, f(x0)
```

```python
#Testando o algoritmo do método de Newton para a função f
import math

x0 = 2.5
max_i = 30

x_otimo, f_otimo = newton(x0, max_i)

print("O valor ótimo foi encontrado em x =", x_otimo, "onde f(x) =", f_otimo)
```
